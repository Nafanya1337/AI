# ЗАДАНИЕ 4.3.1

Задайте массив случайных значений из интервала(0; 1).
Рассчитайте средние и медианные значения для массива, сравните
результаты, какие выводы можно сделать о значениях?
Постройте точечную диаграмму рассения полученного ряда.

```python
import random as rand
import statistics as stat
import matplotlib.pyplot as plt

# создаем список
x = list(rand.uniform(0, 1) for i in range(rand.randint(1, 7)))

# вычисляем среднее и медиану
avg = sum(x)/len(x)
median = stat.median(x)

# построение точечной диаграммы рассеяния
plt.scatter(range(1, len(x) + 1), x)

# добавление линий для среднего и медианы
plt.axhline(avg, color='r', linestyle='-', label='Среднее число')
plt.axhline(median, color='g', linestyle='-', label='Медиана')

# добавление подписей к осям
plt.xlabel("Номер элемента")
plt.ylabel("Значение")

# отображение надписей
plt.legend()
# отображение диаграммы
plt.show()
```

# ЗАДАНИЕ 4.3.2

Дана функция
Создать массив из 10 значений
функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой
половины массива и построить графики для основного массива –
линейный и для среза – точечный

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return pow(1 + math.e ** pow(x, 1/2) + math.cos(x) ** 2, 2) / abs(1 - math.sin(x) ** 3) + math.log(abs(2 * x), math.e)

rand = random.randint(1, 10)
arr = list(func(i) for i in range(rand, 10 + rand))

# выделяем срез
first = arr[:5]

# создаем линейную диаграмму
plt.plot(range(1, len(arr) + 1), arr)

# создаем надписи
plt.ylabel(ylabel='Значение элемента')
plt.xlabel(xlabel='Позиция элемента')

# добавляем точки
plt.scatter(range(1, len(first) + 1), first, color='red')

plt.show()
```

# ЗАДАНИЕ 4.3.3

Дана функция |cos(𝑥𝑒(cos(𝑥)+ln(𝑥+1)))| Построить график на
интервале (0,10) с шагом 1 с заливкой площади и найти эту площадь
под ним. Для вычисления площади воспользуйте функции trapz(),
вычисляющей интеграл по правилу трапеции.
Для ее корректной работы необходимо подключить следующие
библиотеки:
from scipy.integrate import simps
from numpy import trapz

```python
import math
from numpy import trapz
import matplotlib.pyplot as plt

def func(x):
    return abs(math.cos(x * math.e ** (math.cos(x) + math.log(x + 1, math.e))))

arr = list(func(i) for i in range(0, 11))

# создаем линейную диаграмму
plt.plot(range(0, len(arr)), arr)
plt.fill_between(range(0, len(arr)), arr, color='grey')

# создаем надписи
plt.ylabel(ylabel='Значение элемента')
plt.xlabel(xlabel='Позиция элемента')

plt.show()

s = trapz(arr, range(0, 11))
print(s)
```

# ЗАДАНИЕ 4.3.5

Создайте простейший калькулятор, включающий основные действия
для двух переменных ′ + ′, ′ − ′, ′ ∙ ′, ′ ÷ ′, а также вычисление
следующих функций: 𝑒^𝑥+𝑦, sin(𝑥 + 𝑦) , cos(𝑥 + 𝑦) , 𝑥^𝑦.

```python
import math

s = str(input('Введите выражение: '))
znak = ''
if s.find('e') != -1:
    s = s[3:-1]
    s = s.split('+')
    znak = 'e'
elif s.find('sin') != -1:
    s = s[4:-1]
    s = s.split('+')
    znak = 'sin'
elif s.find('cos') != -1:
    s = s[4:-1]
    s = s.split('+')
    znak = 'cos'
elif s.find('+') != -1:
    s = s.split('+')
    znak = '+'
elif s.find('-') != -1:
    s = s.split('-')
    znak = '-'
elif s.find('*') != -1:
    s = s.split('*')
    znak = '*'
elif s.find('/') != -1:
    s = s.split('/')
    znak = '/'
else:
    s = s.split('^')
    znak = '^'
x = float(s[0])
y = float(s[1])
if znak == 'e': print(math.e ** (x + y))
if znak == 'sin': print(math.sin(x + y))
if znak == 'cos': print(math.cos(x + y))
if znak == '^': print(x ** y)
if znak == '+': print(x + y)
if znak == '-': print(x - y)
if znak == '*': print(x * y)
if znak == '/':
    print(x / y) if y != 0 else print('y = 0!')
```
