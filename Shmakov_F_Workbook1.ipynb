–†–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ä–µ–¥–Ω–∏–µ –∏ –º–µ–¥–∏–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è –º–∞—Å—Å–∏–≤–∞, —Å—Ä–∞–≤–Ω–∏—Ç–µ
—Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã, –∫–∞–∫–∏–µ –≤—ã–≤–æ–¥—ã –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –æ –∑–Ω–∞—á–µ–Ω–∏—è—Ö?
–ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Ç–æ—á–µ—á–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ–Ω–∏—è –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞.

```python
import random as rand
import statistics as stat
import matplotlib.pyplot as plt

# —Å–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ–∫
x = list(rand.uniform(0, 1) for i in range(rand.randint(1, 7)))

# –≤—ã—á–∏—Å–ª—è–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∏ –º–µ–¥–∏–∞–Ω—É
avg = sum(x)/len(x)
median = stat.median(x)

# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ç–æ—á–µ—á–Ω–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ—è–Ω–∏—è
plt.scatter(range(1, len(x) + 1), x)

# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –ª–∏–Ω–∏–π –¥–ª—è —Å—Ä–µ–¥–Ω–µ–≥–æ –∏ –º–µ–¥–∏–∞–Ω—ã
plt.axhline(avg, color='r', linestyle='-', label='–°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ')
plt.axhline(median, color='g', linestyle='-', label='–ú–µ–¥–∏–∞–Ω–∞')

# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –ø–æ–¥–ø–∏—Å–µ–π –∫ –æ—Å—è–º
plt.xlabel("–ù–æ–º–µ—Ä —ç–ª–µ–º–µ–Ω—Ç–∞")
plt.ylabel("–ó–Ω–∞—á–µ–Ω–∏–µ")

# –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –Ω–∞–¥–ø–∏—Å–µ–π
plt.legend()
# –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –¥–∏–∞–≥—Ä–∞–º–º—ã
plt.show()
```

–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è
–°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π
—Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π
–ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì
–ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return pow(1 + math.e ** pow(x, 1/2) + math.cos(x) ** 2, 2) / abs(1 - math.sin(x) ** 3) + math.log(abs(2 * x), math.e)

rand = random.randint(1, 10)
arr = list(func(i) for i in range(rand, 10 + rand))

# –≤—ã–¥–µ–ª—è–µ–º —Å—Ä–µ–∑
first = arr[:5]

# —Å–æ–∑–¥–∞–µ–º –ª–∏–Ω–µ–π–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É
plt.plot(range(1, len(arr) + 1), arr)

# —Å–æ–∑–¥–∞–µ–º –Ω–∞–¥–ø–∏—Å–∏
plt.ylabel(ylabel='–ó–Ω–∞—á–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞')
plt.xlabel(xlabel='–ü–æ–∑–∏—Ü–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞')

# –¥–æ–±–∞–≤–ª—è–µ–º —Ç–æ—á–∫–∏
plt.scatter(range(1, len(first) + 1), first, color='red')

plt.show()
```

–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è |cos(ùë•ùëí(cos(ùë•)+ln(ùë•+1)))| –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ –Ω–∞
–∏–Ω—Ç–µ—Ä–≤–∞–ª–µ (0,10) —Å —à–∞–≥–æ–º 1 —Å –∑–∞–ª–∏–≤–∫–æ–π –ø–ª–æ—â–∞–¥–∏ –∏ –Ω–∞–π—Ç–∏ —ç—Ç—É –ø–ª–æ—â–∞–¥—å
–ø–æ–¥ –Ω–∏–º. –î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø–ª–æ—â–∞–¥–∏ –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏–∏ trapz(),
–≤—ã—á–∏—Å–ª—è—é—â–µ–π –∏–Ω—Ç–µ–≥—Ä–∞–ª –ø–æ –ø—Ä–∞–≤–∏–ª—É —Ç—Ä–∞–ø–µ—Ü–∏–∏.
–î–ª—è –µ–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–π —Ä–∞–±–æ—Ç—ã –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–æ–¥–∫–ª—é—á–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ
–±–∏–±–ª–∏–æ—Ç–µ–∫–∏:
from scipy.integrate import simps
from numpy import trapz

```python
import math
from numpy import trapz
import matplotlib.pyplot as plt

def func(x):
    return abs(math.cos(x * math.e ** (math.cos(x) + math.log(x + 1, math.e))))

arr = list(func(i) for i in range(0, 11))

# —Å–æ–∑–¥–∞–µ–º –ª–∏–Ω–µ–π–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É
plt.plot(range(0, len(arr)), arr)
plt.fill_between(range(0, len(arr)), arr, color='grey')

# —Å–æ–∑–¥–∞–µ–º –Ω–∞–¥–ø–∏—Å–∏
plt.ylabel(ylabel='–ó–Ω–∞—á–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞')
plt.xlabel(xlabel='–ü–æ–∑–∏—Ü–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞')

plt.show()

s = trapz(arr, range(0, 11))
print(s)
```

# –ó–ê–î–ê–ù–ò–ï 4.3.5

–°–æ–∑–¥–∞–π—Ç–µ –ø—Ä–æ—Å—Ç–µ–π—à–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä, –≤–∫–ª—é—á–∞—é—â–∏–π –æ—Å–Ω–æ–≤–Ω—ã–µ –¥–µ–π—Å—Ç–≤–∏—è
–¥–ª—è –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö ‚Ä≤ + ‚Ä≤, ‚Ä≤ ‚àí ‚Ä≤, ‚Ä≤ ‚àô ‚Ä≤, ‚Ä≤ √∑ ‚Ä≤, –∞ —Ç–∞–∫–∂–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ
—Å–ª–µ–¥—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π: ùëí^ùë•+ùë¶, sin(ùë• + ùë¶) , cos(ùë• + ùë¶) , ùë•^ùë¶.

```python
import math

s = str(input('–í–≤–µ–¥–∏—Ç–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ: '))
znak = ''
if s.find('e') != -1:
    s = s[3:-1]
    s = s.split('+')
    znak = 'e'
elif s.find('sin') != -1:
    s = s[4:-1]
    s = s.split('+')
    znak = 'sin'
elif s.find('cos') != -1:
    s = s[4:-1]
    s = s.split('+')
    znak = 'cos'
elif s.find('+') != -1:
    s = s.split('+')
    znak = '+'
elif s.find('-') != -1:
    s = s.split('-')
    znak = '-'
elif s.find('*') != -1:
    s = s.split('*')
    znak = '*'
elif s.find('/') != -1:
    s = s.split('/')
    znak = '/'
else:
    s = s.split('^')
    znak = '^'
x = float(s[0])
y = float(s[1])
if znak == 'e': print(math.e ** (x + y))
if znak == 'sin': print(math.sin(x + y))
if znak == 'cos': print(math.cos(x + y))
if znak == '^': print(x ** y)
if znak == '+': print(x + y)
if znak == '-': print(x - y)
if znak == '*': print(x * y)
if znak == '/':
    print(x / y) if y != 0 else print('y = 0!')
```
